```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
from functools import partial
from random import random
from sklearn.preprocessing import MinMaxScaler
from collections import defaultdict
from collections import Counter
from scipy.spatial import distance
from math import sqrt, ceil
import operator
```

```python
df = load_boston()
df.data.shape
```

```python
class kNNRegressioin(object):
    def __init__(self, n_neighbors):
        self.n_neighbors = n_neighbors
    
    def fit(self, X,  y):
        self.X = X
        self.y = y
    
    def calculate(self, u):
        dists = []
        for i in range(len(self.X)):
            dists.append(([y[i]], np.linalg.norm(self.X[i] - u)))
        dists.sort(key=operator.itemgetter(1))
        summ = 0
        i = 0      
        for tup in dists:
            if(i == self.n_neighbors): 
                break
            summ += tup[0][0]
            i += 1  
        return summ/self.n_neighbors
        
    def predict(self, x):
        y_pred = []
        for u in x:
            y_pred.append(self.calculate(u))
        return y_pred
```

```python
scaler = MinMaxScaler()
scaler.fit(df.data)
x = scaler.transform(df.data)
y = df.target

x_test = x[400:]
y_test = y[400:]
x = x[:400]
y = y[:400]
```

```python
n_neighbors = 6
neigh = kNNRegressioin(n_neighbors)
neigh.fit(x, y)
y_pred = neigh.predict(x_test)
for i in range(len(y_pred)):
    print(i+1, ') Predicted value:', round(y_pred[i], 1), '\nReal value:', y_test[i])
```
